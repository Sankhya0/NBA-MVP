### Now we want to compare the MVP players with other players in the NBA in the given time frame

##### To do this we will use the columns player names and the year from both the table and perform a type of natural join on them. But for that we need to make sure that the names are in the same format in both the table. Upon further inspection of the players csv we see that some name have an "*" next to their name which represents players who have been inducted into the basketball hall of fame

```python
players["Player"] = players["Player"].str.replace("*","", regex=False) # replacing the character * with nothing
```

```python
players.head()
```

### Another problem obsereved was the the player names Greg Anderson played for several teams in the year 1991. This will not be usefull for our analysis

```python
players.groupby(["Player","Year"]).get_group(("Greg Anderson",1991))
```

### To counter this we will use the pandas method groupby on the player and year columns so that the stas of a player in a year will be combined together. To do this we will also define a function known as single team

#### As we can see for greg anderson above one TM is labelled as Tot which is the total per game stats of the player in that season, so that problem is solved. However, since that is not a valid team we need to replace the TOT with a valid team. For our analysis wew wwill choose the last team the player played for in that season to be the team we choose for our analysis

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"] # converting the entire dataframe into a single row
        row["Tm"] = df.iloc[-1,:]["Tm"] # choosing the last team the player played for using iloc and assigning that as the team for the returned row
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
```

```python
players
```

#### Group by introduced 2 extra columns namely the player and the year columns (The first 2 columns) so we can get rid of them

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
players.head()
```

```python
players[players["Player"] == "Greg Anderson"]
```

## Now we will combine this combined data with the teams data we obtained previously by web scraping

```python
teams = pd.read_csv("teams.csv")
```

```python
teams.head(10) # we can see a row with central division as the value. We want to remove this from our table
```

```python
teams = teams[~teams["W"].str.contains("Division")].copy() # by doing this we are removing all instances of rows where the "W" column contains the word "Division"
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False) # we will replace the * which represents the teams which reach the playofss with the null
```

#### Another problem we find is that the teams in the teams table and team name in the combined table are different as the names used in the combined table are the nicknames of those teams

```python
sorted(teams["Team"].unique()) # obtaining all the unique teams
```

```python
sorted(combined["Tm"].unique())
```

#### We created a new csv file where the team name were mapped to their nicknames and replaced the nickname with the real name in the combined table. To create the csv just copy and paste the sorted order of team name next to each other.

```python
nicknames = {}
with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]: # skip the header row 
        abbrev,name = line.replace("\n","").split(",") # get rid of the newline character and split the string by ","
        nicknames[abbrev] = name # map the abbreviation to the complete team name
```

```python

combined["Team"] = combined["Tm"].map(nicknames)
```

```python
combined.head() # we see that the Team now shows the complete name as opposed to the nickname
```

#### Now since the team name as are in the same format in both the tables, we can merge the 2 table based on the Year and Team Name

```python
train = combined.merge(teams, how="outer",on=["Team", "Year"])
```

#### As we can see many of the columns in the above table still have the data type as string which will not be useful in the machine learning stage. To deal with this wwe will use the inbuilt pandas pd.to_numeric function to convert those data types to int and float as required

```python
train = train.apply(pd.to_numeric, errors='ignore') 
```
